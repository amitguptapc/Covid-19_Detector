### 2. Visualizing the X-ray Scans

```
import matplotlib.pyplot as plt
import cv2
plt.style.use('seaborn')
```

```
pos = cv2.imread('./CovidDataset/Train/Covid/paving.jpg')
plt.imshow(pos)
plt.axis("off")
plt.title('Covid +ve X-ray')
plt.show()
print('\n')
neg = cv2.imread('./CovidDataset/Train/Normal/IM-0300-0001.jpeg')
plt.imshow(neg)
plt.axis("off")
plt.title('Covid -ve X-ray')
plt.show()
```
